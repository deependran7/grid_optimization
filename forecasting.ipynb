```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd
```

```python
xlabel = np.array([2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022])
xdata = np.arange(len(xlabel))
ydata = np.array([515.24,557.53,603.28,648.39,721.73,812.5,885.28,946.1,1026.65, 1095, 1201, 1291, 1385, 1444, 1508, 1320, 1408, 1482, 1748 ])
```

```python
xlabel = np.arange(2003,2045,step=1)
x_forecast = np.arange(len(xlabel))
```

```python
def func(x, a, b):
    return a*np.exp(b*x)
```

```python
popt, pcov = curve_fit(func, xdata, ydata)
```

```python
xdataf = np.arange(len(x_forecast))
ydataf = func(xdataf, *popt)
```

```python
load_data = pd.DataFrame()
load_data['year'] = pd.date_range(start='2003', end='2045',freq='Y')
xdataf = np.arange(len(load_data))
load_data['Forecasted Data'] = ydataf
load_data.loc[range(len(ydata)),'Historical Data'] = ydata
load_data = load_data.set_index('year')
```

```python
fig,ax = plt.subplots(figsize = (12,6))
load_data.plot(ax = ax)
ax.set_ylabel('Peak Load (MW')
ax.grid()
```
